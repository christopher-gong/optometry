```python
roi_df['binaryPax'] = (roi_df.Pax6 > 1500)*1
roi_df['binaryParv'] = (roi_df.Parv > 800)*1
roi_df
```

```python
#drop non Pax, non Parv data
roi_df2 = roi_df[['binaryPax','binaryParv']]
roi_df2[0:10]
```

```python
print("Both binaryPax and binaryParv: " + str(len(roi_df2[(roi_df2['binaryPax'] == 1) & (roi_df2['binaryParv'] == 1)])))
print("Only binaryPax: " + str(len(roi_df2[(roi_df2['binaryPax'] == 1) & (roi_df2['binaryParv'] == 0)])))
print("Only binaryParv: " + str(len(roi_df2[(roi_df2['binaryPax'] == 0) & (roi_df2['binaryParv'] == 1)])))
print("Niether binaryPax nor binaryParv: " + str(len(roi_df2[(roi_df2['binaryPax'] == 0) & (roi_df2['binaryParv'] == 0)])))
```

```python
# produce the clustergram using the dataframe that we created above
net.load_df(roi_df2)
# normalize our data
net.normalize(axis='column', norm_type='zscore', keep_orig=True)
#runs clustering with default parameters. LOOK INTO PARAMETERS - default is linkage type average
net.cluster()
net.widget()
```
