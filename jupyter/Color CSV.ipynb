```python
def colorCSV(df, col, colors, csvLoc):
    results = []
    for c in df[col]:
        results += [colors[int(c)]]
    df = pd.DataFrame(data={'COLORS': results})
    df.to_csv(csvLoc, index=False, header=False)
    return df
```

```python
colorCSV(roi_df, '0', ["red", "green", "yellow", "blue", "orange"], "colored.csv")
```
